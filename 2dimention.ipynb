# PCA在二维上的演示

参考 http://www.dscademy.com/unsupervised-learning/principal-component-analysis/index.php?from=header

```python
import numpy as np
from sklearn import decomposition
from matplotlib import pyplot as plt
```

```python
# 创建1000个数据，分布在0.3X上的点
xs = np.arange(-5,5, 0.01)
def y_gen(x):
    return(np.random.normal(0.3*x, np.sqrt(25 - abs(x)**2)/5))
ys = list(map(y_gen, xs))
X = np.c_[(xs, ys)]
```

```python
# 创建计算主成分分析的对象
pca_computer = decomposition.PCA(n_components=1)
# 拟合主成分分析对象
pca_computer.fit(X)
# 使用此对象对原数据进行转换（新投影）
X_transformed = pca_computer.transform(X)
```

```python
# 查看转换后数据的形状
X_transformed.shape
```

```python
# 查看主成分
pca_computer.components_ # 可以注意0位置为1位置0.3倍，且两者的平方和为1
```

```python
# 将降维后的数据投影回去
X_projected = X_transformed.dot(pca_computer.components_)
```

```python
# 作图表示
plt.plot(X[:, 0], X[:, 1], '.', label='original data')
plt.plot(X_projected[:, 0], X_projected[:, 1], '.', label='projected data')
plt.xlim((-7, 7))
plt.ylim((-7, 7))
plt.legend()
```
